```python
import pandas as pd
data= pd.read_csv('framingham.csv')
data.head()
```

üßç Demographics & Lifestyle
Column Name	Meaning
male:(Binary): 1 if the patient is male, 0 if female.
age:Age of the patient (in years).
education:Education level (usually encoded numerically from 1 to 4).
currentSmoker:(Binary): 1 if currently smokes, 0 if not.
cigsPerDay:	Average number of cigarettes smoked per day.

ü©∫ Medical History & Conditions
Column Name	Meaning
BPMeds	(Binary): 1 if on blood pressure medication, 0 if not.
prevalentStroke	(Binary): 1 if the patient had a stroke before, 0 if not.
prevalentHyp(Binary): 1 if diagnosed with hypertension previously, 0 if not.
diabetes(Binary): 1 if has diabetes, 0 if not.

üß™ Biometrics & Lab Values
Column Name	Meaning
totChol:Total cholesterol level (mg/dL).
sysBP:Systolic blood pressure (mm Hg).
diaBP:Diastolic blood pressure (mm Hg).
BMI:Body Mass Index (weight/height¬≤).
heartRate:Resting heart rate (beats per minute).
glucose:Blood glucose level (mg/dL).

TenYearCHD:If the patient would develop a heart disease on 10 years

```python
data=data.drop('education',axis=1)
```

```python
data
```

```python
data = data.reset_index(drop=True)
```

```python
X=data.drop('TenYearCHD',axis=1)
Y=data['TenYearCHD']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression(max_iter=40000)
```

```python
classifier.fit(X_train,Y_train)
predictions1= classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score= accuracy_score(Y_test,predictions1)
score
```

```python
import joblib
joblib.dump(classifier, "logistic_model_framingham.joblib")
```

```python
import matplotlib.pyplot as plt


# Assuming you have column names in X
feature_names = X.columns
coefficients = classifier.coef_[0]

# Create DataFrame
coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
}).sort_values(by='Coefficient', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(coef_df['Feature'], coef_df['Coefficient'], color='teal')
plt.xlabel('Coefficient Value')
plt.title('Feature Importance in Logistic Regression')
plt.axvline(0, color='black', linewidth=0.8)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

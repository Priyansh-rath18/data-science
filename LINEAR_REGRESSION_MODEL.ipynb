```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error,
```

```python
data=pd.read_csv('Salary_dataset.csv')
```

```python
data
```

```python
data = data.drop(columns=['Unnamed: 0'])
```

```python
data
```

```python
X= data[['YearsExperience']]
Y= data['Salary']
model= LinearRegression()
model.fit(X,Y)
```

```python
y_pred= model.predict(X)
```

```python
experience = float(input("Enter years of experience: "))
predicted_salary = model.predict([[experience]])
print(f"Predicted Salary: ₹{predicted_salary[0]:,.2f}")

```

```python
import joblib
```

```python
joblib.dump(model, 'salary_predictor.joblib')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt

mae = mean_absolute_error(Y, y_pred)
mse = mean_squared_error(Y, y_pred)
rmse = sqrt(mse)
r2 = r2_score(Y, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```

#MAE (Mean Absolute Error): 4644.20

On average,  model's salary predictions are ₹4644.20 off from the actual salaries.

This is a straightforward and interpretable metric.

#MSE (Mean Squared Error): 31,270,951.72

Squared average of the errors. It's more sensitive to outliers than MAE.

High value due to squaring (₹ is squared, so units are not intuitive).

#RMSE (Root Mean Squared Error): 5592.04

This is in the same unit as salary (₹), and tells you the typical deviation from the actual salary.

So predictions are, on average, ₹5592.04 off (more robust than MAE if there are outliers).

#R² Score: 0.96

This means 96% of the variance in salary is explained by model using years of experience.

This is an excellent score, indicating your model fits the data very well.

```python
import numpy as np
import matplotlib.pyplot as plt

# Continuous uniform from 0 to 1
samples = np.random.uniform(0, 1, 10000)

plt.hist(samples, bins=50, density=True, alpha=0.6, color='skyblue')
plt.title("Continuous Uniform Distribution (0 to 1)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

IN THE ABOVE CODE 10_000 RANDOM VALUES ARE CREATED BETWEEN 0 AND 1,These are sampled from a continuous uniform distribution, where:

Every value in the range [0,1] is equally likely.

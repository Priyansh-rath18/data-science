```python
import requests
```

```python
url=("https://quotes.toscrape.com/")
response=requests.get(url)
```

```python
print(response.text)
```

```python
print(response.status_code)
```

```python
if response.status_code == 200:
    print("Page fetched successfully!")
else:
    print("Failed to fetch the page.")
```

```python
with open("htmls/fetchedpage.html",'w',encoding='utf-8') as f:
    f.write(response.text)
```

```python
for i in range(1,50):
    response=requests.get(f"https://quotes.toscrape.com/page/{i}/")
    with open(f"htmls/fetchedpage{i}.html",'w',encoding='utf-8') as f:
        f.write(response.text)
    print(f"page{i} downloaded successfully")
```

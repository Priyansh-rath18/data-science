```python
import pandas as pd
data=pd.read_csv('Indian_Kids_Screen_Time.csv')
data
```

```python
data.dropna
```

```python
data.isnull().sum()
```

```python
data['Health_Impacts'].unique()
```

```python
data['Health_Impacts'] = data['Health_Impacts'].fillna('No Symptoms')

data.reset_index(drop=True)
```

```python
data['Health_Impacts'] = data['Health_Impacts'].apply(lambda x: 1 if x == 'Poor Sleep' else 0 if x == 'No Symptoms' else 2)
```

```python
data
```

```python
data['Age_Group'] = data['Age'].apply(lambda x: 0 if 8 <= x <= 13 else 1 if 14 <= x <= 18 else None)
```

```python
data
```

```python
data['Gender'].unique()
```

```python
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})
```

```python
data
```

```python
data['Exceeded_Recommended_Limit'].unique()
```

```python
data['Exceeded_Recommended_Limit'] = data['Exceeded_Recommended_Limit'].apply(lambda x: 1 if x else 0)
data
```

```python
data.drop('Age',axis=1)
```

```python
data['Primary_Device'].unique()
```

```python
data['Primary_Device'] = data['Primary_Device'].apply(lambda x: 1 if x == 'Laptop' else 0 if x == 'Smartphone' else 2 if x== 'TV' else 3 )
data
```

```python
data['Urban_or_Rural'].unique()	
```

```python
data['Urban_or_Rural']=data['Urban_or_Rural'].map({'Urban':1,'Rural':0})
data
```

```python
data.drop('Age',axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Features (all columns except 'Health_Impacts')
X = data.drop(columns=['Health_Impacts'])

# Target column
Y = data['Health_Impacts']

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
new_data = [[12, 1, 4.5, 1, 1, 0.36, 2, 1]]  # List in the same order as X_train columns
prediction = model.predict(new_data)
print("Predicted Health Impact:", prediction[0])

```
